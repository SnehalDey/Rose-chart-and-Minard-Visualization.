```python
#plotting the figure showing the advancement and retreat and the thickness of the line represnt the number of survivors. 
troops_chart = alt.Chart(troops).mark_trail().encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    size=alt.Size(
        'Survivors', 
        scale=alt.Scale(range=[1, 35]), 
        legend=None
    ),
    detail='Division',
    color=alt.Color(
        'Direction', 
        scale=alt.Scale(
            domain=['A', 'R'],
            range=['#9fc9b8', '#d42f19']
        ), 
        legend=None
    ),
).project(
    type="mercator"
)
troops_chart=troops_chart.properties(
 height=200,width=900)
```

```python
troops_chart
```

```python
encodeX = alt.X(
    'Longitude:Q',
    scale=alt.Scale(
        domain=[city["Longitude"].min(), city["Longitude"].max()]
    ),
    axis=None
)

encodeY = alt.Y(
    'Temperature',
    axis=alt.Axis(
        title="Temperature During The Retreat", 
        grid=True, 
        orient='right'
    )
)

```

```python
# drawing the temperature graph during the reterat
temp_chart = alt.Chart(temperatures).mark_line(
    color="#888888"
).encode(
    x=encodeX,
    y=encodeY,
) + alt.Chart(temperatures).mark_text( 
    dy=10,
    font='sans-serif', 
    fontSize=10
).encode(
    x=encodeX,
    y=encodeY,
    text='label'
)

temp_chart = temp_chart.properties(
    height=200,width=900
)
```
